```python
from livelossplot import PlotLosses
import torch as ch
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
import pandas as pd
import torch.nn as nn
import seaborn as sns
from train_target import get_loaders, get_model_and_criterion, load_data
from scipy import stats

import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 300
```

```python
# CIFAR
num_train_points = 25000
num_pop_points = 25000 #10000
batch_size = 64
learning_rate = 0.001
epochs = 100

dataset = "CIFAR10"
device = "cuda"
```

```python
# Use same subset of data across all experiments
all_data, train_index, test_index = load_data(num_train_points, num_pop_points)
```

```python
criterion = nn.CrossEntropyLoss()
```

```python
def compute_gradients_on_data(model, loader):
    # Set model to eval mode
    model.eval()

    # Compute gradients of model on all given data
    running_gradients = []
    for x, y in loader:
        x, y = x.to(device), y.to(device)
        model.zero_grad()
        y_hat = model(x)
        loss = criterion(y_hat, y)
        loss.backward()
        flat_grads = ch.cat(
            [p.grad.detach().cpu().flatten() for p in model.parameters()]
        )
        running_gradients.append(flat_grads)
    # Compute mean gradient (across data, per param)
    mean_grads = ch.stack(running_gradients).mean(dim=0)

    # Zero out model gradients now
    model.zero_grad()

    return mean_grads
```

```python
def compute_grad_noise_statistics(model, population_grads, loader):
    # Set model to eval mode
    model.eval()
    
    # Compute gradients of model on all given data
    m1, m2 = ch.zeros_like(population_grads), ch.zeros_like(population_grads)
    batches = 0
    for x, y in loader:
        x, y = x.to(device), y.to(device)
        model.zero_grad()
        y_hat = model(x)
        loss = criterion(y_hat, y)
        loss.backward()
        flat_grads = ch.cat(
            [p.grad.detach().cpu().flatten() for p in model.parameters()]
        )
        grad_noise = population_grads - flat_grads
        m1 += grad_noise
        m2 += grad_noise ** 2
        batches += 1

    m1 /= batches
    m2 /= batches
    return m1, m2
```

```python
@ch.no_grad()
def collect_loss_values(model, loader):
    criterion_each = nn.CrossEntropyLoss(reduction='none')
    # Set model to eval model
    model.eval()
    
    all_losses = []
    for x, y in loader:
        x, y = x.to(device), y.to(device)
        y_hat = model(x).detach()
        losses = criterion_each(y_hat, y)
        all_losses.append(losses.cpu())
    return ch.cat(all_losses)
```

```python
def model_training_trial():
    # Get loaders (use same ordering of data across model inits)
    train_loader, test_loader = get_loaders(all_data, train_index, test_index,
                                            batch_size, start_seed=2023)

    model, criterion = get_model_and_criterion(dataset, device=device)
    
    # Set the loss function and optimizer
    optimizer = ch.optim.Adam(model.parameters(), lr=learning_rate)
    
    groups = {'acc': ['acc'], 'Trace (not normalized)': ['Trace (not normalized)'], 'loss': ['loss']}
    plotter = PlotLosses(groups=groups)

    # Loop over each epoch
    trace_values, losses, accuracies = [], [], []
    m1_vals, m2_vals = [], []
    mem_losses, nonmem_losses = [], []
    for epoch_idx in range(epochs):
        train_loss = 0
        train_acc = 0
        samples_seen = 0
    
        # Loop over the training set
        for data, target in train_loader:    
            # Make sure model is in train mode
            model.train()

            # Move data to the device
            data_use = data.to(device, non_blocking=True)
            target = target.to(device, non_blocking=True)
            samples_seen += len(target)

            # Set the gradients to zero
            optimizer.zero_grad(set_to_none=True)

            # Get the model output
            output = model(data_use)

            # Calculate the loss
            binary_case = False
            if output.shape[1] == 1:
                # BCE loss case
                binary_case = True
                loss = criterion(output.squeeze(), target.float())
            else:
                # CE loss case
                loss = criterion(output, target.long())

            # Perform the backward pass
            loss.backward()

            # Take a step using optimizer
            optimizer.step()

            # Add the loss to the total loss
            train_loss += loss.detach().cpu().item()

            # Computing accuracy
            with ch.no_grad():
                if binary_case:
                    pred = output.data.squeeze() > 0.5
                    train_acc += pred.eq(target.data.view_as(pred)).sum().cpu().item()
                else:
                    pred = output.data.max(1, keepdim=True)[1]
                    train_acc += pred.eq(target.data.view_as(pred)).sum().cpu().item()

        # Compute mean gradients on population data
        mean_grads = compute_gradients_on_data(model, test_loader)
        
        m1, m2 = compute_grad_noise_statistics(model, mean_grads, train_loader)

        # Scale with learning rate
        trace_value = ch.sum(1 / (m2 - m1 **2)) * (2 / learning_rate)

        # Save iter-wise values
        trace_values.append(trace_value.cpu().item())
        losses.append(loss.cpu().item())
        accuracies.append(100 * train_acc / samples_seen)
        m1_vals.append(m1.detach().cpu().numpy())
        m2_vals.append(m2.detach().cpu().numpy())
        
        # Also keep track of loss values for members and nonmembers
        # mem_losses.append(collect_loss_values(model, train_loader).numpy())
        # nonmem_losses.append(collect_loss_values(model, test_loader).numpy())
    
        # Plot on livelossplot
        plotter.update(
            {
                "loss": loss.cpu().item(),
                "acc": 100 * train_acc / samples_seen,
                "Trace (not normalized)": trace_value.item()
            }
        )
        plotter.send()

        print(f"Epoch: {epoch_idx+1}/{epochs}")

    return trace_values, losses, accuracies, mem_losses, nonmem_losses, m1_vals, m2_vals
```

```python
all_trace_values, all_losses, all_accuracies, mem_losses, nonmem_losses = [], [], [], [], []
all_m1vals, all_m2vals = [], []

n_trials = 5
for _ in range(n_trials):
    returned_values = model_training_trial()
    all_trace_values.append(returned_values[0])
    all_losses.append(returned_values[1])
    all_accuracies.append(returned_values[2])
    mem_losses.append(returned_values[3])
    nonmem_losses.append(returned_values[4])
    all_m1vals.append(returned_values[5])
    all_m2vals.append(returned_values[6])
```
